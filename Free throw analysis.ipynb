```python
# Create the plotting function
def myplot(data, playerlist = Players):

    Col = {"KobeBryant":"Black","JoeJohnson":"Red","LeBronJames":"Green","CarmeloAnthony":"Blue","DwightHoward":"Magenta","ChrisBosh":"Black","ChrisPaul":"Red","KevinDurant":"Green","DerrickRose":"Blue","DwayneWade":"Magenta"}
    Mrk = {"KobeBryant":"s","JoeJohnson":"o","LeBronJames":"^","CarmeloAnthony":"D","DwightHoward":"s","ChrisBosh":"o","ChrisPaul":"^","KevinDurant":"D","DerrickRose":"s","DwayneWade":"o"}

    for name in playerlist:
        plt.plot(data[Pdict[name]],c=Col[name],ls='--',marker=Mrk[name],ms='7',label=name)
    plt.legend(loc='upper left',bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)),Seasons,rotation='vertical')
    plt.show()
```

```python
myplot(FreeThrows)
```

```python
myplot(FreeThrowAttempts)
```

Once the basic matrix have been created, and after checking them with a plot, we can deliver interesting insights. Let's plot the number of attemps per game

```python
#Free Throw Attempts Per Game 
myplot(FreeThrowAttempts/Games)
```

FTA per Game varies during the career of all the players. There is a decreasing tendency in the total amount of attemps per games in all the players since 2005 except for Dwight Howard and Kevin Durant which is similar to 2005 (altought it has fallen compared to other years with higher ratios). Joe Johnson gets few attempts per game

```python
myplot(FreeThrows/FreeThrowAttempts)
```

At first glance all the players, except for Dwight Howard, had high accuracy in free throws. Derrick Rose and Chris Paul had the highest improvements. Chris Paul had the highest accuracy despite being getting fewer attempts.

```python
#Player playing style excluding free throws
myplot((Points - FreeThrows) / FieldGoals)
```

Field Goals can only be for either 2 points or 3 points. This is verified in this plot.

We can see that almost all players experiment with their style throughout their careers. The most drastic change in style has been experienced by Joe Johnson, evolving to a three line player. On the other side, Dwight Howard was a two point score expert, playing close to the basket, during all these years.
